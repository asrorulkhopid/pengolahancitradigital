```python
# import library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# membaca dan menampilkan image
image=plt.imread("modul1.jpg")
plt.imshow(image,"gray")
```

```python
# fungsi untuk mengubah setiap pixel RGB ke gray dengan mencari nilai rata-rata tiap R,G, dan B sebagai nilai gray
def RGB2GRAY(image):
    gray= np.zeros(image.shape).astype(int) # membuat sebuah numpy array nol sebagai wadah untuk menyimpan hasil grayscale
    for i in range (gray.shape[0]):
        for j in range (gray.shape[1]):
                gray[i,j]=(int(image[i][j][0])+int(image[i][j][1])+int(image[i][j][2]))/3 # mencari nilai gray untuk pixel
    return gray # mengembalikan nilai
```

```python
gray=RGB2GRAY(image)
```

```python
plt.imshow(gray)
```
