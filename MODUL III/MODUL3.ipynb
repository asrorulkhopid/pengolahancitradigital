```python
#import library yang di butuhkan
import cv2 # pengolahan image
import matplotlib.pyplot as plt # gambar dan histogram
import numpy as np # membantu dalam perhitungan angka atau bilangn tertentu
```

```python
image=cv2.imread("modul3_1.jpg") # baris code ini digunakan untuk membaca gambar dan disimpan dalam sebuah variabel (image)
plt.title("IMAGE RGB")  # baris code ini memanggil function title() dari libary pyplot yang berfungsi 
                        # untuk memberikan judul pada gambar yang akan ditampilkan
plt.imshow(image) # baris code ini memanggil function imshow() dari libary pyplot yang berfungsi
                  # untuk menampilkan image yang diberikan sebagai input dari function tersebut 
```

```python
grayImage=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) # berfungsi untuk mengubah warna gambar dari RGB ke Gray
plt.title("GRAY IMAGE") # baris code ini berfungsi untuk memberikan judul pada gambar yang akan ditampilkan nanti
plt.imshow(grayImage,"gray") # baris code ini berfungsi untuk menampilkan gambar yang diinputkan dengan colormap gray
```

```python
# fungsi berikut digunakan untuk membuat histogram dari image (sebagai input)
def histogram(Image,title):
    keabuan=np.arange(Image.max()+1) # baris code ini berfungsi untuk membuat sebuah np array yang nilainya dari 0 sampai nilai
                                     # maksimal yang ada pada array image
    ni=np.zeros(Image.shape[0]) # baris code ini berfungsi untuk membuat sebuah np array nol yang nantinya akan digunakan untuk
                                # menyimpan banyak-nya kemunculan untuk setiap nilai keabuan pada image
    # perulangan untuk menghitung kemunculan tiap nilai keabuan
    for i in range (Image.shape[0]): # melakukan perulangan sebanyak baris untuk menelusuri setiap baris image
        for j in range (Image.shape[0]): # melakukan perulangan sebanyak kolom untuk menelusuri setiap kolom pada image
            ni[Image[i][j]]+=1 # increment pada indeks ni yang nilai warnanya ditemukan pada pixel
    plt.title(title) # berfungsi untuk memberikan judul pada histogram yang akan ditampilkan
    plt.xlabel("Keabuan") # memberikan keterangan pada sumbu x histogram
    plt.ylabel("Kemunculan") # memberikan keterangan pada sumbu y histogram
    plt.bar(keabuan,ni) # membuat histogram dengan data keabuan dan ni
```

```python
histogram(grayImage,"HISTOGRAM GRAY IMAGE")
print(grayImage.max())
```

```python
  # fungsi berikut untuk melakukan equalisasi pada histogram untuk image
def equalize(img):
    ni = np.zeros(img.max()+1).astype(int) # membuat sebuah np array nol dengan ukuran sebanyak nilai max pada image
                                             # dengan tipe data untuk tiap elemen pada array adalah  integer
    tinggi, lebar = img.shape       # berfungsi untuk mengambil u

    # perulangan untuk menghitung kemunculan tiap tiap keabuan
    for i in range(tinggi ): # perulangan sebanyak tinggi untuk menelusuri setiap baris pada image
        for j in range(lebar): # perulangan sebanyak lebar untuk menelusuri setiap kolom pada image
            ni [img[i,j]] +=1 # menghitung kemunculan keabuan dengan melakukan increment pada ni dengan indeks sesuai keabuan yang muncul

    # menentukan kemunculan kumulatif
    kumulatif = np.zeros(img.max()+1).astype(int) # membaut sebuah np array nol dengan ukuran sesuai nilai maximal pada image
    for i in range(img.max()+1): # melakukan perulangan sebanyak nilai maksimal atau sesuai ukuran dengan array kumulatif
        kumulatif[i] = np.sum ( ni[0:i+1] ) # menghitung nilai kumulatif 
    
    # menentukan tujuan pemetaan
    target_value = np.floor(kumulatif*(img.max())/(tinggi*lebar)).astype(int)
    # fungsi tersebut berfungsi untuk mencari sekaligus mendefinisikan tujuan pemetaan untuk setiap pixel pada image

    # melakukan proses pemetaan untuk setiap pixel
    result = np.zeros(img.shape) #variabel yang berisi np array kosong seukuran image yang berfungsi untuk menyimpan hasil pemetaan tiap pixel
    for i in range(tinggi): # perulangan sebanyak tinggi image untuk menelusuri setiap baris pada image
        for j in range(lebar): # perulangan sebanyak lebar image untuk menelusuri setiap kolom pada image
            result[i, j] = target_value[img[i,j]] # pemetaan
 
    return result # mengembalikan nilai / output dari fungsi ini adalah result
```

```python
hasil=equalize(grayImage).astype(int) # memanggil fungsi equalize dengan input berupa grayImage
```

```python
# menampilkan image hasil equalisasi
plt.title("IMAGE HASIL EQUALISASI HISTOGRAM") # memberikan judul pada image yang akan di tampilkan
plt.imshow(hasil,"gray") # menampilkan image (hasil) dengan colormap Gray
```

```python
histogram(hasil,"EQUALISASI HISTOGRAM GRAY IMAGE")
```

```python
# SPESIFIKASI
plt.figure(figsize=(10,5))
image2 = cv2.imread("modul3_2.jpg") # baris code ini digunakan untuk membaca gambar dan menggunakan library openCV
grayImage2=cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY) # mengubah warna dari RGB ke Gray
plt.subplot(121)
plt.imshow(grayImage2,"gray") # menampilkan gambar (grayImage2) dengan colormap gray
plt.subplot(122)
histogram(grayImage2,"HISTOGRAM IMAGE 2")
```

```python
# method spesifikasi tahap 1 ini adalah method untuk melakukan equalisasi pada gambar-gambar yang akan di spesifikasi dan target gambar
def spesifikasi_tahap1(img):
    ni = np.zeros(img.max()+1).astype(int) # membuat sebuah np array nol dengan ukuran sebanyak nilai max pada image
                                             # dengan tipe data untuk tiap elemen pada array adalah  integer
    tinggi, lebar = img.shape       # berfungsi untuk mengambil u

    # perulangan untuk menghitung kemunculan tiap tiap keabuan
    for i in range(tinggi ): # perulangan sebanyak tinggi untuk menelusuri setiap baris pada image
        for j in range(lebar): # perulangan sebanyak lebar untuk menelusuri setiap kolom pada image
            ni [img[i,j]] +=1 # menghitung kemunculan keabuan dengan melakukan increment pada ni dengan indeks sesuai keabuan yang muncul

    # menentukan kemunculan kumulatif
    kumulatif = np.zeros(img.max()+1).astype(int) # membaut sebuah np array nol dengan ukuran sesuai nilai maximal pada image
    for i in range(img.max()+1): # melakukan perulangan sebanyak nilai maksimal atau sesuai ukuran dengan array kumulatif
        kumulatif[i] = np.sum ( ni[0:i+1] ) # menghitung nilai kumulatif 
    
    # menentukan tujuan pemetaan
    target_value = np.floor(kumulatif*(img.max())/(tinggi*lebar)).astype(int)
    # fungsi tersebut berfungsi untuk mencari sekaligus mendefinisikan tujuan pemetaan untuk setiap pixel pada image

    return target_value
```

```python
# memanggil method untuk mendapatkan nilai yang digunakan untuk spesikasi yaitu taget1(gambar yang di spesifikasi) dan gambar2(tujuan spesifikasi)
target1=spesifikasi_tahap1(grayImage)
target2=spesifikasi_tahap1(grayImage2)
```

```python
# melakukan pemetaan untuk setiap nilai pada target1 ke nilai terdekat dari target2
def pemetaan(img,target1,target2):
    pemetaan = np.zeros(img.max()+1).astype(int) # membuat sebuah numpy array nol untuk menyimpan tujuan pemetaan
    k=0 # variabel bantu untuk mempermudah penyimpanan nilai pemetaan
    warna_terdekat=0 # deklarasi sementara untuk warna terdekat
#     perulangan untuk mencari nilai terdekat dari target1 ke target2
    for i in (target1):
        min = 1000 # deklarasi sementara untuk min
        for j in range(len(target2)): # perulangan sepanjang ukuran target2
            if (abs(i-target2[j]))<min : # melakukan pengecekan untuk mengambil dan menemukan nilai paling dekat
                min = (abs (i-target2[j])) # menggunakan selisih absolut untuk menemukan nilai terdekat
                                           # semakin kecil nilai absolut dari selisih dua buah bilangan maka semakin dekat bilangan tersebut
                warna_terdekat=j # mengupdate indeks nilai terdekat
        pemetaan[k]=warna_terdekat #menyimpannya dalam wadah yang sudah disiapkan
        k+=1 # increment untuk nilai k
    return pemetaan # mengembalikan nilai
```

```python
pemetaan=pemetaan(grayImage,target1,target2) # memanggil function yang sudah di definisikan
```

```python
imageSpesifikasi = np.zeros(grayImage.shape).astype(int) # membuat variabel yang berisi numpy array nol untuk menyimpan hasil dari spesifikasi 
# porulangan untuk mengunjungi setiap pixel pada image
for i in range(grayImage.shape[0]): #
    for j in range(grayImage.shape[1]):
        imageSpesifikasi [i][j] = pemetaan[grayImage[i][j]] #memetakan setiap pixel ke alamat yang sudah disiapkan
```

```python
#menampilkan gambar dan histogram
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.title("Image 1 Hasil Spesifikasi")
plt.imshow(imageSpesifikasi,"gray")
plt.subplot(222)
plt.title("Image 2 Referensi/Target")
plt.imshow(grayImage2,"gray")
plt.subplot(223)
plt.title("Hist 1 Hasil Spesifikasi")
histogram(imageSpesifikasi,"HISTOGRAM HASIL SPESIFIKASI")
plt.subplot(224)
plt.title("Hist 2 Referensi/Target")
histogram(imageSpesifikasi,"HISTOGRAM TARGET")
```
